```python
import SimpleITK as sitk
import os

input_root = r"C:\Users\ardra nirwan\COCA_data\cocacoronarycalciumandchestcts-2\Gated_release_final\patient"
output_root = r"C:\Users\ardra nirwan\COCA_nifti"

os.makedirs(output_root, exist_ok=True)

for patient in os.listdir(input_root):
    patient_path = os.path.join(input_root, patient)

    if not os.path.isdir(patient_path):
        continue

    for scan in os.listdir(patient_path):
        scan_path = os.path.join(patient_path, scan)

        if not os.path.isdir(scan_path):
            continue

        reader = sitk.ImageSeriesReader()
        series_ids = reader.GetGDCMSeriesIDs(scan_path)

        if not series_ids:
            print(f"Skipping {scan_path}")
            continue

        series_files = reader.GetGDCMSeriesFileNames(scan_path, series_ids[0])
        reader.SetFileNames(series_files)

        image = reader.Execute()

        output_path = os.path.join(output_root, f"patient_{patient}.nii")
        sitk.WriteImage(image, output_path)

        print(f"Done patient {patient}")

print("ALL DONE ")
```

```python
## Step 1: Resampling
import os
import SimpleITK as sitk

input_folder = r"C:\Users\ardra nirwan\COCA_nifti"
output_folder = r"C:\Users\ardra nirwan\COCA_resampled"

os.makedirs(output_folder, exist_ok=True)

def resample_image(image, new_spacing=[1.0, 1.0, 1.0]):
    original_spacing = image.GetSpacing()
    original_size = image.GetSize()

    new_size = [
        int(round(original_size[i] * (original_spacing[i] / new_spacing[i])))
        for i in range(3)
    ]

    resampler = sitk.ResampleImageFilter()
    resampler.SetOutputSpacing(new_spacing)
    resampler.SetSize(new_size)
    resampler.SetOutputDirection(image.GetDirection())
    resampler.SetOutputOrigin(image.GetOrigin())
    resampler.SetInterpolator(sitk.sitkLinear)

    return resampler.Execute(image)

count = 0

for file in os.listdir(input_folder):
    if file.endswith(".nii"):
        input_path = os.path.join(input_folder, file)
        output_path = os.path.join(output_folder, file)

        image = sitk.ReadImage(input_path)
        resampled = resample_image(image)

        sitk.WriteImage(resampled, output_path)

        count += 1
        print(f"Resampled {file}")

print(f"ALL DONE  Total: {count}")
```

```python
import os
import SimpleITK as sitk
import numpy as np

input_folder = r"C:\Users\ardra nirwan\COCA_resampled"
output_folder = r"C:\Users\ardra nirwan\COCA_windowed"

os.makedirs(output_folder, exist_ok=True)

def apply_window(image, center=300, width=600):
    img_array = sitk.GetArrayFromImage(image)

    lower = center - width // 2
    upper = center + width // 2

    img_array = np.clip(img_array, lower, upper)

    # normalize to 0–1
    img_array = (img_array - lower) / (upper - lower)

    new_image = sitk.GetImageFromArray(img_array)
    new_image.CopyInformation(image)

    return new_image

count = 0

for file in os.listdir(input_folder):
    if file.endswith(".nii"):
        input_path = os.path.join(input_folder, file)
        output_path = os.path.join(output_folder, file)

        image = sitk.ReadImage(input_path)
        windowed = apply_window(image)

        sitk.WriteImage(windowed, output_path)

        count += 1
        print(f"Windowed {file}")

print(f"ALL DONE  Total: {count}")
```

```python
import os
import random
import shutil

input_folder = r"C:\Users\ardra nirwan\COCA_windowed"

base_output = r"C:\Users\ardra nirwan\COCA_split"
train_dir = os.path.join(base_output, "train")
val_dir = os.path.join(base_output, "val")
test_dir = os.path.join(base_output, "test")

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

files = [f for f in os.listdir(input_folder) if f.endswith(".nii")]
random.shuffle(files)

total = len(files)
train_split = int(0.7 * total)
val_split = int(0.85 * total)

train_files = files[:train_split]
val_files = files[train_split:val_split]
test_files = files[val_split:]

def copy_files(file_list, destination):
    for f in file_list:
        shutil.copy2(os.path.join(input_folder, f), os.path.join(destination, f))

copy_files(train_files, train_dir)
copy_files(val_files, val_dir)
copy_files(test_files, test_dir)

print(f"Train: {len(train_files)}")
print(f"Val: {len(val_files)}")
print(f"Test: {len(test_files)}")
print("SPLIT DONE ")
```

```python
#dataloader

class COCADataset:
    def __init__(self, folder):
        self.files = [os.path.join(folder, f) for f in os.listdir(folder) if f.endswith(".nii")]

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        file_path = self.files[idx]
        image = nib.load(file_path).get_fdata()
        return image
```

```python
import SimpleITK as sitk
import os

dicom_folder = r"C:\Users\ardra nirwan\COCA_data"
output_folder = r"C:\Users\ardra nirwan\COCA_nifti"

os.makedirs(output_folder, exist_ok=True)

reader = sitk.ImageSeriesReader()

for root, dirs, files in os.walk(dicom_folder):
    try:
        series_IDs = reader.GetGDCMSeriesIDs(root)
        if not series_IDs:
            continue

        for i, series_id in enumerate(series_IDs):
            series_file_names = reader.GetGDCMSeriesFileNames(root, series_id)
            reader.SetFileNames(series_file_names)

            image = reader.Execute()

            folder_name = os.path.basename(root)
            output_name = f"{folder_name}_series{i}.nii.gz"
            output_path = os.path.join(output_folder, output_name)

            sitk.WriteImage(image, output_path)

            print(f"Converted: {output_name}")

    except Exception as e:
        print(f"Error in {root}: {e}")
```

```python
import os
import pandas as pd
from radiomics import featureextractor

image_dir = r"C:\Users\ardra nirwan\COCA_nifti"
mask_dir = r"C:\Users\ardra nirwan\COCA_output"

extractor = featureextractor.RadiomicsFeatureExtractor()

data = []

for folder in os.listdir(mask_dir):
    mask_path = os.path.join(mask_dir, folder, "heart.nii.gz")

    
    image_path = None
    for img in os.listdir(image_dir):
        if folder in img:
            image_path = os.path.join(image_dir, img)
            break

    if not os.path.exists(mask_path) or image_path is None:
        print(f"Skipping: {folder}")
        continue

    print(f"Processing: {folder}")

    try:
        features = extractor.execute(image_path, mask_path)
        
        features_dict = dict(features)
        features_dict["patient"] = folder
        
        data.append(features_dict)

    except Exception as e:
        print(f"Error in {folder}: {e}")

df = pd.DataFrame(data)

output_csv = r"C:\Users\ardra nirwan\radiomics_features.csv"
df.to_csv(output_csv, index=False)

print("DONE Saved at:", output_csv)
```
